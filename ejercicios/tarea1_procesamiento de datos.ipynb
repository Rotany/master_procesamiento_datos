##  Variables Númericas

Discretas : Son valores que estan representados por números enteros

- Son: age, trestbps, chol, thalach

Continuas : Son valores estan representados por números decimales


- Son : oldepeak

## Variables Categoricas

Ordinal: Una variable puede ser tratada como ordinal cuando sus valores representan categorías con alguna clasificación intrínseca

- Son: ca, restcg 

Nominal : Cuando sus valores representan categorías 

- Son:  sex, fbs, cp, exang, slope, num, thal
